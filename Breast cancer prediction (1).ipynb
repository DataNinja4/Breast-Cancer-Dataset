```python
data["diagnosis"] = data["diagnosis"].astype("category", copy=False)
```

```python
data["diagnosis"].value_counts().plot(kind = "bar")
```

```python
# Let's start building our model
```

```python
X = data.drop(["diagnosis"], axis = 1)
y = data["diagnosis"]
```

```python
# Our featiures vary differentlt because they are different measurements and different units
# Let's standardize the features/variables to help our model know which feature has the most weight
```

```python
from sklearn.preprocessing import StandardScaler

# Create a standard Scaler
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# Now, Let's split and train and test our data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
# Predict the target variable on test data
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"The Accuracy of our model is: {accuracy: .2f}")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
